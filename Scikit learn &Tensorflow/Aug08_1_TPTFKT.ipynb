```python
sentence = """영국에서 흉기난동 참사 이후 반(反)이민·반무슬림 극우 시위가 이어지는 가운데 7일(현지시간) 이에 맞선 반극우 '맞불' 시위가 영국 전역에서 벌어졌다.
이날 저녁 런던과 브라이튼, 리버풀, 버밍엄, 브리스틀 등지에서는 이민자 지원 센터 앞을 비롯한 거리 곳곳에 수천명이 모인 가운데 인종주의와 극우에 반대하는 시위가 열렸다.
BBC·스카이방송, 일간 가디언 등에 따르면 시위대는 "난민을 환영한다", "인종주의를 거부한다"고 쓰인 플래카드를 들고 구호를 외쳤다."""
sentence
```

```python
sentence2 = sentence.replace(" ", "") # 띄어쓰기 없애기
sentence2
```

```python
from pykospacing import Spacing
```

```python
spacing = Spacing()
s = spacing(sentence2)

print(sentence)
print()
print(s)
```

```python
# SNS나 채팅의 경우 'ㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋ', 'ㅎㅎㅎㅎㅎㅎㅎㅎㅎㅎㅎㅎㅎ' 같이
# 불필요하게 연속되는 경우가 꽤나 많은데
# 반복되는 것들을 하나로 정규화시켜줄 수 있음
print(emoticon_normalize('ㅋㅋㅋㅋㅋㅋㅋㅋㅋ앜ㅋㅋㅋㅋㅋㅋ', num_repeats=2))
print(emoticon_normalize('ㅎㅎㅎㅎㅎㅎㅎㅎㅎ앜ㅋㅋㅋㅋㅋㅋ', num_repeats=2))
print(emoticon_normalize('ㅠㅠㅠㅠㅠㅠㅠㅠㅠ앜ㅋㅋㅋㅋㅋㅋ', num_repeats=2))
print(emoticon_normalize('ㅜㅜㅜ앜ㅋㅋㅋㅋㅋㅋ', num_repeats=2))
```

```python
print(emoticon_normalize('으하하하하하하하하하하하핫', num_repeats=3))
print(emoticon_normalize('으하하하하하하하하하하하핫', num_repeats=2))
print(emoticon_normalize('으히히히히히히히히히히힣', num_repeats=5))
```
