# 명령어

- Ctrl + Enter : 해당 셀을 실행
- Shift + Enter : 해당 셀을 실행하고 다음 셀로 이동
- Ctrl + M + Z : 실행 취소
- Ctrl + M + D : 해당 셀을 지우기
- Ctrl + :     : 텍스트 셀 만들기
- 도구 탭 => 단축키
